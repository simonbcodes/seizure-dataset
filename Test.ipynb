```python
import pandas as pd
import numpy as np

from bokeh.plotting import figure, show, output_notebook
from bokeh.palettes import Spectral6, Spectral3
from bokeh.transform import linear_cmap
from bokeh.models import ColumnDataSource, HoverTool

output_notebook()
```

```python
height = 100
p = figure(x_range=(0, height), y_range=(0, height))
p.axis.visible = False
p.xgrid.visible = False
p.ygrid.visible = False
p.toolbar.logo = None
p.toolbar_location = None
p.grid.minor_grid_line_color = 'white'

grid = np.random.randint(0, height, 100)

mapper = linear_cmap(field_name='x', palette=Spectral6, low=0, high=height)

while True:
    source = ColumnDataSource(dict(x=np.random.randint(0, height, 100), y=np.random.randint(0, height, 100)))
    boi = p.circle(x='x', y='y', line_color='black', color=mapper, line_width=0.5, alpha=0.4, source=source, size=20, hover_fill_color='white')
    p.add_tools(HoverTool(tooltips = None, renderers=[boi]))
    show(p)
```
